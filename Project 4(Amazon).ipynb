# Q9: average rating and rating count for products

```python
data['rating'] = pd.to_numeric(data['rating'], errors='coerce')
data['rating_count'] = pd.to_numeric(data['rating_count'], errors='coerce')
```

```python
# Calculate the average rating
average_rating = data['rating'].mean()

# Calculate the total rating count for all products
total_rating_count = data['rating_count'].sum()

# Calculate the average rating count per product
average_rating_count_per_product = total_rating_count / len(data)

print(f"Average Rating: {average_rating:.2f}")
print(f"Average Rating Count per Product: {average_rating_count_per_product:.2f}")
```

# Q10: distribution of products across different categories

```python
# Split the sub-categories from the 'category' column
data['category'] = data['category'].apply(lambda x: x.split('|')[0])

# Count the number of products in each category
product_count_by_category = data['category'].value_counts()

# Plot the distribution using a bar chart
plt.figure(figsize=(10, 6))
product_count_by_category.plot(kind='bar', color='red')
plt.xlabel('Product Category')
plt.ylabel('Number of Products')
plt.title('Distribution of Products Across Different Categories')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Q11: correlation between ratings, rating count, and discounted price

```python
data['rating'] = pd.to_numeric(data['rating'], errors='coerce')
data['rating_count'] = pd.to_numeric(data['rating_count'], errors='coerce')
data['discounted_price'] = pd.to_numeric(data['discounted_price'], errors='coerce')
```

```python
correlation = data[['rating', 'rating_count', 'discounted_price']].corr()

print(correlation)
```

# Q12: top-rated products with a high number of ratings

```python
# Sort the DataFrame by 'rating' in descending order and then by 'rating_count' in descending order
top_rated_products = data.sort_values(by=['rating', 'rating_count'], ascending=[False, False])

# Display the top-rated products with a high number of ratings
print(top_rated_products[[ 'product_name', 'rating', 'rating_count']])
```

# Q13: distribution of products by discount percentage

```python
dp = data.groupby(["discount_percentage"],as_index= False)["product_name"].count()
```

```python
plt.figure(figsize=(19, 8))
sns.barplot(data = dp, x= dp["discount_percentage"],y=dp["product_name"],color = "Gray")
plt.title("Distribution of product based on discount percentages")
plt.xticks(rotation=45)
plt.xlabel("Discount_percentage")
plt.ylabel("Count of product")
plt.show()
```

# Q14: average discount percentage by category

```python
data['discount_percentage'] = pd.to_numeric(data['discount_percentage'], errors='coerce')
# Split the sub-categories from the 'category' column
data['category'] = data['category'].apply(lambda x: x.split('|')[0])

# Calculate the average discount percentage by category
average_discount_by_category = data.groupby('category')['discount_percentage'].mean()

print(average_discount_by_category)
```
